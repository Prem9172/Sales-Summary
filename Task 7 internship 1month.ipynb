```python
import sqlite3

# Create a new SQLite database (or connect if exists)
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()
```

```python
# Create a simple 'sales' table
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    product TEXT,
    quantity INTEGER,
    price REAL
)
''')
```

```python
# Insert some sample data
sample_data = [
    ('Apples', 10, 2.5),
    ('Bananas', 15, 1.5),
    ('Apples', 5, 2.5),
    ('Oranges', 20, 3.0),
    ('Bananas', 10, 1.5),
    ('Grapes', 8, 4.0),
    ('Oranges', 5, 3.0),
    ('Grapes', 7, 4.0)
]
```

```python
cursor.executemany("INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)", sample_data)
conn.commit()
```

```python
import pandas as pd

query = '''
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue 
FROM sales 
GROUP BY product
'''
```

```python
df = pd.read_sql_query(query, conn)
```

```python
print("ðŸ“¦ Sales Summary by Product:")
print(df)
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar', x='product', y='revenue', color='skyblue', legend=False)
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue (â‚¹)')
plt.tight_layout()
plt.savefig("sales_chart.png")  # Save chart to file
plt.show()
```
